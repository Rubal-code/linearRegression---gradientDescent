```python
import numpy as np
from sklearn.datasets import make_regression
```

```python
X,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

**lets apply ols (ordinary least square)**

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='r')
plt.show()
```

```python
# now lets apply gradient descent lets suppose our slope is 78.35
# i am going to assume the value for b (itercept) initially =0
y_pred=((78.35 * X) + 0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='r',label='OLS')
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
m=78.35
b=0
loss_slope= -2 * np.sum(y - m*X.ravel() - b)          # .ravel takes 2d,3d,____ arrays and return it into 1d array it is just like .flatten()
loss_slope
```

```python
# lets taking learning rate =0.1
n=0.1
step_size=n*loss_slope
step_size
```

```python
# calculating the new intercept
b=b-step_size
```

```python
b      # 0--> 20.92  appa 26.15 tak jana
```

```python
y_pred1=((m*X)+b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='r',label='OLS')
plt.plot(X,y_pred,color='green',label='b=0')
plt.plot(X,y_pred1,color='black',label='b={}'.format(b))
plt.legend()
plt.show()
```

```python
# iteration 2
loss_slope=-2*np.sum(y-m*X.ravel()-b)
```

```python
loss_slope
```

```python
step_size=n*loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred2=((m*X)+b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='r',label='OLS')
plt.plot(X,y_pred,color='green',label='b=0')
plt.plot(X,y_pred1,color='black',label='b={}'.format(b))
plt.plot(X,y_pred2,color='orange',label='b={}'.format(b))
plt.legend()
plt.show()
```

```python
#iteration3
loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
step_size=n*loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred3=((m*X+b)).reshape(4)
y_pred3
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='r',label='OLS')
plt.plot(X,y_pred,color='green',label='b=0')
plt.plot(X,y_pred1,color='black',label='b={}'.format(b))
plt.plot(X,y_pred2,color='orange',label='b={}'.format(b))
plt.plot(X,y_pred3,color='purple',label='b={}'.format(b))
plt.legend()
plt.show()
```

```python
b=100 # now i suppose intercept 100 intially
m=78.35
n=0.01  # now learning rate i take 0.01 means not big steps

epochs=100  # iterations
for i in range(epochs):
    loss_slope=-2 * np.sum(y-m*X.ravel()-b)
    step_size=n*loss_slope
    b=b-step_size
    y_pred=((m*X)+b).reshape(4)
    plt.plot(X,y_pred)
plt.scatter(X,y)    

```
